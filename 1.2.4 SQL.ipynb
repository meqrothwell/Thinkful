What are the three longest trips on rainy days?

SELECT
	t.trip_id,
	t.duration
from
	trips t
join
	weather w
ON
	t.zip_code = w.ZIP
where
	w.Events like 'Rain'
Order by t.duration desc
limit 3


----

SELECT
	t.trip_id,
	max(t.duration)
from
	trips t
join
	weather w
ON
	t.zip_code = w.ZIP
where
	w.Events like 'Rain'
group by 1
Order by t.duration desc



Which station is empty most often?

with
	locations
as (
	select
		station_id,
		count(bikes_available) ct
	from 
		status
	where
		bikes_available = 0
)

SELECT
	stations.station_id,
	stations.name,
	max(locations.ct) empty
from
	stations
join
	locations
ON
	stations.station_id = locations.station_id

Return a list of stations with a count of number of trips start at that station but ordered by dock count.

select
	t.start_station,
	count(*) trip_counts,
	s.dockcount
from	
	trips t
join
	stations s
on 
	t.start_station = s.name
group by
	t.start_station
order by
	s.dockcount desc

WITH 
	rainy
as (
	select
		date
	from 
		weather
	where 
		Events like 'Rain'
	group by 1
),
rain_trips
as (
	SELECT 
		trip_id,
		duration,
		date(start_date) trip_date
	from
		trips
	join
		rainy
	on
		rainy.date = trip_date
	order by
		duration desc
)

select
	trip_date,
	max(duration)
from
	rain_trips
group by 1
