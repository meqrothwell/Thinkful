```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Normal Distribution

```python
normal_sample = np.random.normal(50, 2, 100)
```

```python
plt.hist(normal_sample)
plt.show()
```

```python
plt.hist(normal_sample)
plt.axvline(normal_sample.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normal_sample.mean() + normal_sample.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(normal_sample.mean() - normal_sample.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
```

The mean and standard deviation are valuable in descibing a normal distribution.

# Zipf Distribution

```python
zipf_sample = np.random.zipf(3, 100)
plt.hist(zipf_sample)
plt.show()
```

```python
plt.hist(zipf_sample)
plt.axvline(zipf_sample.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(zipf_sample.mean() + zipf_sample.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(zipf_sample.mean() - zipf_sample.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
```

# Dirichlet Distribution

```python
dirichlet_sample = np.random.dirichlet([1, 1], 100)
plt.hist(dirichlet_sample)
plt.show()
```

```python
plt.hist(dirichlet_sample)
plt.axvline(dirichlet_sample.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(dirichlet_sample.mean() + dirichlet_sample.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(dirichlet_sample.mean() - dirichlet_sample.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
```

# Normal Distribution Addition

```python
normal_1 = np.random.normal(5, .5, 10000)
normal_2 = np.random.normal(10, 1, 10000)
plt.hist(normal_1)
plt.hist(normal_2)
plt.show()
```

```python
normal_3 = normal_1 + normal_2
plt.hist(normal_3)
plt.show()
```

```python
plt.hist(normal_3)
plt.axvline(normal_3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normal_3.mean() + normal_3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(normal_3.mean() - normal_3.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
```

When adding two datasets with a normal distribution, the resulting data set's mean will be the sum of the two original means and the standard deviation appears to be just over 1 (or the greater of the two original standard deviations).
