```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans
%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/llimllib/bostonmarathon/master/results/2014/results.csv')
data.head()
```

```python
data['male'] = data['gender'] == 'M'
data = data.dropna(axis=1, how='any')
data = data.dropna(axis=0, how='any')
```

```python
data.describe()
```

```python
data.head()
```

```python
data_nums = data[['division', 'age', 'official', 'genderdiv', 'overall', 'pace', 'male']]
X = data_nums.iloc[:, 0:6]
y = data_nums.iloc[:, 6]
```

```python
plt.hist(data['age']);
```

```python
sns.swarmplot(x="age", y="pace", hue="male", data=data)
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(data_nums, alpha=0.2, figsize=(6, 6), diagonal='kde');
```

```python
# Normalize the data.
X_norm = normalize(X)

# Reduce it to five components.
X_pca = PCA(5).fit_transform(X_norm)

# Calculate predicted values.
y_pred = KMeans(n_clusters=3, random_state=42).fit_predict(X_pca)

# Plot the solution.
plt.scatter(X_pca[:, 0], X_norm[:, 1], c=y_pred)
plt.show()

# Check the solution against the data.
print('Comparing k-means clusters against the data:')
print(pd.crosstab(y_pred, y))
```

```python
# Get predicted clusters.
full_pred = KMeans(n_clusters=3, random_state=42).fit_predict(X_norm)

pd.crosstab(y, full_pred) 
```

```python
from sklearn import metrics
    
metrics.adjusted_rand_score(y, full_pred)
```

```python
# Each batch will be made up of 200 data points.
minibatchkmeans = MiniBatchKMeans(
    init='random',
    n_clusters=3,
    batch_size=200)
minibatchkmeans.fit(X_pca)

# Add the new predicted cluster memberships to the data frame.
predict_mini = minibatchkmeans.predict(X_pca)

# Check the MiniBatch model against our earlier one.
print('Comparing k-means and mini batch k-means solutions:')
print(pd.crosstab(predict_mini, y_pred))
```

```python
from sklearn import metrics
    
metrics.adjusted_rand_score(predict_mini, y_pred)
```

```python
#Divide into training and test sets.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.9,
    random_state=42)
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

# Here we set the bandwidth. This function automatically derives a bandwidth
# number based on an inspection of the distances among points in the data.
bandwidth = estimate_bandwidth(X_train, quantile=0.05, n_samples=500)

# Declare and fit the model.
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(X_train)

# Extract cluster assignments for each data point.
labels = ms.labels_

# Coordinates of the cluster centers.
cluster_centers = ms.cluster_centers_

# Count our clusters.
n_clusters_ = len(np.unique(labels))

print("Number of estimated clusters: {}".format(n_clusters_))
```

```python
#plt.scatter(X_train[:, 0], X_train[:, 1], c=labels)
#plt.show()

print('Comparing the assigned categories to the ones in the data:')
print(pd.crosstab(y_train,labels))
```

```python
from sklearn import metrics
    
metrics.adjusted_rand_score(y_train, labels)
```

```python
from sklearn.cluster import SpectralClustering

n_clusters=5

# Declare and fit the model.
sc = SpectralClustering(n_clusters=n_clusters)
sc.fit(X_train)

#Predicted clusters.
predict=sc.fit_predict(X_train)

#Graph results.
plt.scatter(X_train['age'], X_train['pace'], c=predict)
plt.show()

print('Comparing the assigned categories to the ones in the data:')
print(pd.crosstab(y_train,predict))
```

## Conclusion

For this data set, Mini Batch K-Means yielded the best results, with and Adjusted Rand Score of 0.984, meaning that there was nearly perfect agreement between the ground truth and the new solution. Normal K-Means also worked reasonably well, with and ARI of 0.812. Mean Shift worked very poorly, with an ARI of 0.11, meaning it was slightly better than random. Using Spectral Clustering, the best result I could get was just slightly negative but essentially 0, meaning that the algorithm resulted in no improvement over simply guessing randomly.
